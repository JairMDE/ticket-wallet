```python
import pandas as pd
import numpy as np 

import cv2
import matplotlib.pylab as plt
```

# Testing Prototype

```python
img_path = '/Users/jairmartinez/Documents/IoT Project/IoT-doctr/images/ticket-1.jpeg'
original_img = cv2.imread(img_path)
```

```python
gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# Gaussian Blur 
blurred_img = cv2.GaussianBlur(gray_img, (5,5), 0)

#Tresholding to get a binary image
_, binary_img = cv2.threshold(blurred_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1, 4, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1, 4, 3)
plt.imshow(blurred_img, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(1, 4, 4)
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
```

## Applying an adaptive treshold method

```python
adaptive_trsh_image = cv2.adaptiveThreshold(blurred_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap="gray")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(adaptive_trsh_image, cmap='gray')
plt.title('Adaptive Threshold Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
